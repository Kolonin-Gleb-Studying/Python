```python
from matplotlib import pyplot as plt
from random import randint

```

```python
x = [* range(1, 21)] # * - Распаковка списка
y = [randint(1, 100) for i in x]
y1 = [randint(1, 70) for i in x]
print(x)
print(y)
```

```python
plt.figure(figsize=(10, 5)) #Размер поля для графика
plt.title("График") #Название графика
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
#Обозначаем на осях нужные нам значения
plt.xticks(x)

plt.plot(x, y, 'r:') #Рисуем по координатам из списка
plt.plot(x, y1, 'b*--')
plt.show()
```

```python
# Самостоятельная y = x^2

x = [* range(-10, 11)]
y = [i**2 for i in x]
print(x)
print(y)

plt.figure(figsize=(10, 10))
plt.title("График параболы")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
#Обозначаем на осях нужные нам значения
plt.xticks(x)


plt.plot(x, y, 'r') #Рисуем по координатам из списка
plt.show()

```

```python
#Another code block

import numpy as np

x = [*range(18, 30)]
index_x = np.arange(len(x))


y1 = [randint(1000, 10000) for i in x]
y2 = [randint(1000, 10000) for i in x]
y3 = [randint(1000, 10000) for i in x]


plt.figure(figsize=(10, 5))
plt.title("Столбчатая диаграмма")
plt.bar(index_x, y1, width=0.25)
plt.bar(index_x+0.25, y2, width=0.25) # Делаем смещение столбцов, чтобы они не накладывались
plt.bar(index_x+0.5, y3, width=0.25)
plt.xticks(ticks=index_x, labels=x) #Проставляем значения x по оси х
plt.legend(['1', '2', '3'])
plt.show()

```
